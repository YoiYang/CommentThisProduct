```python
import json
import pandas as pd

    
df = pd.read_json('reviews_Musical_Instruments_5.json', lines=True)
    
```

```python
df
```

```python
good_reviews = df[df['overall'] > 3]['reviewText']
bad_reviews = df[df['overall'] <= 3]['reviewText']
```

```python
Infile = open('goodreviews.txt','w')
Infile2 = open('badreviews.txt','w')

for line in good_reviews:
    Infile.write(line + '\n')
    
    
for line in bad_reviews:
    Infile2.write(line + '\n')
    
Infile.close()
Infile2.close()
```
